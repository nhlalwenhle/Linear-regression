```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Import salary.csv as data using pd.read_csv method and use .dropna() to drop all nan values.
data = pd.read_csv("salary.csv").dropna()
```

```python
y = data.salary
x = data.yearsworked
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state = 170)
```

```python
model = sm.OLS(y_train, X_train).fit()
```

```python
model_test = sm.OLS(y_test, X_test).fit()
```

```python
model_test.summary()
```

```python
model.summary()
```

```python
y, x = data.salary, data.yearsworked
beta, beta0, r_value, p_value, std_err = stats.linregress(x,y)
print("y = %f x + %f, r: %f, r-squared: %f,\np-value: %f, std_err: %f"
% (beta, beta0, r_value, r_value**2, p_value, std_err))
print("Regression line with the scatterplot")
yhat = beta * x + beta0 # regression line
plt.plot(x, yhat, 'r-', x, y,'o')
plt.xlabel('yearsworked (Experience)')
plt.ylabel('Salary (Dollars)')
plt.show()
print("Using seaborn")
import seaborn as sns
sns.regplot(x="yearsworked", y="salary", data=data);
```
